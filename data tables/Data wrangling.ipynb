# Data Wrangling

```python
import pandas as pd
```

```python
#import covariate datasets qnd the outcomes. 

cancer_path = '/Users/alecstear/Documents/GitHub/anti-diabetics-study-with-EHR/data tables/cancer_diab_pop.csv'
df_cancer = pd.read_csv(cancer_path)

age_sex_path = '/Users/alecstear/Documents/GitHub/anti-diabetics-study-with-EHR/data tables/age_sex_diab_pop.csv'
df_age_sex = pd.read_csv(age_sex_path)

COPD_path = '/Users/alecstear/Documents/GitHub/anti-diabetics-study-with-EHR/data tables/COPD_diab_pop.csv'
df_COPD = pd.read_csv(COPD_path)

diabetes_cohort_path = '/Users/alecstear/Documents/GitHub/anti-diabetics-study-with-EHR/data tables/diab_pop.csv'
df_diabetes_cohort = pd.read_csv(diabetes_cohort_path)

exposure_drug_prescription_path = '/Users/alecstear/Documents/GitHub/anti-diabetics-study-with-EHR/data tables/exposure_drugs.csv'
df_exp_drug = pd.read_csv(exposure_drug_prescription_path)

fracture_path = '/Users/alecstear/Documents/GitHub/anti-diabetics-study-with-EHR/data tables/fracture_diab_pop.csv'
df_fracture = pd.read_csv(fracture_path)

hypertension_path = '/Users/alecstear/Documents/GitHub/anti-diabetics-study-with-EHR/data tables/hypertension_diab_pop.csv'
df_hypertension = pd.read_csv(hypertension_path)

surgery_path = '/Users/alecstear/Documents/GitHub/anti-diabetics-study-with-EHR/data tables/surgery_diab_pop.csv'
df_surgery = pd.read_csv(surgery_path)

outcome_VTE_path = '/Users/alecstear/Documents/GitHub/anti-diabetics-study-with-EHR/data tables/outcome_VTE.csv'
df_outcome = pd.read_csv(outcome_VTE_path)

#add event date table : outcome_VTE
```

```python
df_outcome
```

```python
df_diabetes_cohort = df_diabetes_cohort.rename(columns={'pat_id': 'PAT_ID'})
df_outcome = df_outcome.rename(columns={'pat_id': 'PAT_ID'})
total_cohort_df = pd.merge(df_diabetes_cohort, df_age_sex, on='PAT_ID', how='inner')
df_diabetes_cohort = total_cohort_df
```

```python
df_diabetes_cohort['cancer'] = df_diabetes_cohort['PAT_ID'].isin(df_cancer['PAT_ID']).astype(int)
df_diabetes_cohort['COPD'] = df_diabetes_cohort['PAT_ID'].isin(df_COPD['PAT_ID']).astype(int)
df_diabetes_cohort['fracture'] = df_diabetes_cohort['PAT_ID'].isin(df_fracture['PAT_ID']).astype(int)
df_diabetes_cohort['hypertension'] = df_diabetes_cohort['PAT_ID'].isin(df_hypertension['PAT_ID']).astype(int)
df_diabetes_cohort['surgery'] = df_diabetes_cohort['PAT_ID'].isin(df_surgery['PAT_ID']).astype(int)
print(df_diabetes_cohort['cancer'].sum(), df_diabetes_cohort['COPD'].sum(), df_diabetes_cohort['fracture'].sum(), df_diabetes_cohort['hypertension'].sum(), df_diabetes_cohort['surgery'].sum())
```

```python
counter = 0
for i in df_diabetes_cohort['PAT_ID']:
    # Check if 'pat_id' is in 'PAT_ID' column of df_outcome
    if i in df_outcome['PAT_ID'].values:
        counter += 1
        # Update 'prs_date' in df_diabetes_cohort based on the condition
        df_diabetes_cohort.loc[df_diabetes_cohort['PAT_ID'] == i, 'eventdate outcome'] = df_outcome.loc[df_exp_drug['PAT_ID'] == i, 'eventdate'].values[0]
counter
```

```python
counter = 0
for i in df_diabetes_cohort['PAT_ID']:
    # Check if 'pat_id' is in 'PAT_ID' column of df_exp_drug
    if i in df_exp_drug['PAT_ID'].values:
        counter += 1
        # Update 'prs_date' in df_diabetes_cohort based on the condition
        df_diabetes_cohort.loc[df_diabetes_cohort['PAT_ID'] == i, 'prs_date'] = df_exp_drug.loc[df_exp_drug['PAT_ID'] == i, 'prs_date'].values[0]
counter
```
